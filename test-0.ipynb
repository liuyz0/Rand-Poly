Try to find roots of polynomials.

```julia
using Polynomials, Plots
```

```julia
p = Polynomial(randn(200))
r = roots(p);
```

```julia
scatter(real(r),imag(r))
```

```julia
real(r[isreal.(r)])
```

```julia
p.(real(r[isreal.(r)]))
```

```julia
import AMRVW; const AM = AMRVW

r1 = AM.roots(coeffs(p))
```

```julia
scatter(real(r1),imag(r1))
```

```julia
real(r1[isreal.(r1)])
```
